# 価値反復
## 確率ロボティクス　課題　18C1015 岩井一輝

### 取り組んだ内容
5×5の簡単な世界を準備し, ロボットが1マス1歩で移動できると設定して, 何歩でゴールすることができるかを求める。しかし, この世界には障害物があり, ロボットは障害物にぶつからないように慎重に移動するため, 障害物の周りでは5歩余計に必要であるとする。また, このロボットは右＞上＞左＞下の優先度で行動しようとする。

```python
%matplotlib inline
import numpy as np

size = 5                                                    
costs = [[10,10,10,10,10],[10,10,10,10,10],[10,50,10,10,10],[10,50,10,10,10],[10,10,10,10,0]]   #ゴールのコストを0, 障害物のコストを50(任意)と設定        
goal = (4,4)   
obstacle = (2,1) #障害物の座標指定
actions = ["up","down","left","right"] # ロボットが選択できる行動のリスト

def cost(pos): # 指定した座標のコストを返す補助の関数
    p = [pos[0],pos[1]]
   #壁にぶつかったらそのまま 
    for i in [0,1]:  
        if p[i] < 0:        p[i] = 0
        if p[i] >= size: p[i] = size - 1
    
            
    return costs[p[0]][p[1]]
```

```python
import matplotlib.pyplot as plt  
import matplotlib.patches as patches

def draw(mark_pos, action):  # mark_pos はマス目に目印の枠をつけるときに使います
    fig, ax = plt.subplots()
    mp = ax.pcolor(costs, cmap=plt.cm.YlOrRd,vmin=0,vmax=8)
    ax.set_aspect(1)
    ax.set_xticks(range(size), minor=False)
    ax.set_yticks(range(size), minor=False)
    
    for x in range(len(costs)):
        for y in range(len(costs[0])):
            plt.text(x+0.5,y+0.5,costs[x][y],ha = 'center', va = 'center', size=20)
            
    plt.text(goal[0]+0.75,goal[1]+0.75,"G",ha = 'center', va = 'center', size=20)
    plt.text(obstacle[0]+0.75,obstacle[1]+0.75,"O",ha = 'center', va = 'center', size=20)
    
    if mark_pos != None:
        plt.text(mark_pos[0]+0.5,mark_pos[1]+0.25,action,ha = 'center', va = 'center', size=20)
            
    plt.show()
    fig.clear()
    
draw(None,None)
```

```python
def action_value(pos, action, goal, obstacle):
    if pos == goal: return cost(pos)    # ゴールの価値は不変
    if pos == obstacle: return cost(pos)    #障害物の価値も固定
    if    action == "up":
        return 1+cost((pos[0],pos[1]+1))
    elif action == "down": return 1+cost((pos[0],pos[1]-1))
    elif action == "left":    return 1+cost((pos[0]-1,pos[1]))
    elif action == "right":  return 1+cost((pos[0]+1,pos[1]))
```

```python
def sweep():
    changed = False
    
    for x in range(size):
        for y in range(size):
            min_cost = 100
            min_action = None
            for a in actions:
                c = action_value((x,y), a, goal, obstacle)
                if c < min_cost:
                    min_cost = c
                    min_action = a
            
            if costs[x][y] != min_cost:
                costs[x][y] = min_cost
                draw((x,y),min_action)
                changed = True
            
            #障害物が隣にある場合コストを上げる
            
    return changed

draw(None,None)
changed = True
n = 1
while changed:
    print("sweep",n)
    changed = sweep()
    n = n + 1
    
draw(None,None)
```

```python
#ゴールからの距離を表すコストに障害物を避けるように障害物に隣接する座標のコストを引き上げる
draw(None,None)
for x in range(size):
    for y in range(size):
        if (x,y) == obstacle:
            costs[x][y] -= 5
        for s in [-1, 0, 1]:
            for t in [-1, 0, 1]:
                if (x+s,y+t) == obstacle:
                    costs[x][y] += 5
draw(None,None)
```

```python
#計算したコストからロボットの行動を決定
#ロボットはright > up > left > down の優先度で行動を選択する
x = 0
y = 0
while((x,y) != goal):
    if x < size - 1 and costs[x][y] > costs[x+1][y]:
        x += 1
        print('right')
    elif  y < size -1 and costs[x][y] > costs[x][y+1]:
        y += 1
        print('up')
    elif x > 0 and costs[x][y] > costs[x-1][y]:
        x -= 1
        print('left')
    elif y > 0 and costs[x][y] > costs[x][y-1]:
        y -= 1
        print('back')
        
```
